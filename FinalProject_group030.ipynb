# **Data Cleaning**
The data that comes from the API as a json is generally unclean. The json that is exported for each hashtag included extraneous columns such as “effectStickers” and has certain important data in an unreadable format such as the like count of the TikTok being in the “stats” column in a dictionary within a variable called “diggCount”. In order to clean this data to be more understandable and easier to work with, it is important that we drop the columns that we do not need and extract the numbers that we do need. 


1.   Retrieve columns with the like number (diggCount), comment number (commentCount), and play count (playCount) for renaming purposes
2.   Remove columns with information on the creator, video IDs, and any unnecessary columns that we don't use and/or violate privacy
3.   Remove nulls
4.   Remove punctuaions in the video descriptions and lowercase them 


```
# returns like count from dictionary
def retrieveLikesCount(dict):
    retValue = dict['diggCount']
    return retValue

# returns comment count from dictionary
def retrieveCommentCount(dict):
    retValue = dict['commentCount']
    return retValue

# returns view count from dictionary
def retrieveViewCount(dict):
    retValue = dict['playCount']
    return retValue

# returns a lowercase-stripped version of a string
def descToLowerStrip(string):
    retValue = string.lower()
    return retValue.strip()

# removes punctuation from the string
def removePunctuation(string):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~’'''
    for letter in string:
        if letter in punctuations:
            string = string.replace(letter, "") 
    return string

# drops extraneous colunmns from the dataframe, retrieve like, comment and view count, and make the description lowercase, strips description and removes punctuation
def reformDataFrame(dataframe):
    df_temp = dataframe.drop(columns=['author','id','challenges','createTime','digged','duetEnabled','forFriend','shareEnabled','showNotPass','stickersOnItem','stitchEnabled','textExtra','video','vl1','warnInfo','effectStickers','isAd','itemCommentStatus','itemMute','music','officalItem','originalItem','privateItem','secret','duetInfo'])
    df_temp["likesCount"] = df_temp["stats"].apply(retrieveLikesCount)
    df_temp["commentCount"] = df_temp["stats"].apply(retrieveCommentCount)
    df_temp["viewCount"] = df_temp["stats"].apply(retrieveViewCount)
    df_temp = df_temp.drop(columns=['authorStats','stats'])
    df_temp['desc'] = df_temp['desc'].apply(descToLowerStrip)
    df_temp['desc'] = df_temp['desc'].apply(removePunctuation)
    return df_temp
```

Now that we've created our functions to clean our data, we can clean our dataframes to include relevant data pertaining to our TikToks

# **Data Analysis & Results**

Using data collected on the frequency of certain political hashtags (such as **#trump2020** and **#biden2020**) and their respective political biases, we strove to create a predictive model that could analyze the descriptions of a set of hashtags and determine whether or not they were right or left leaning surrounding the election of 2020 in November.

Our data analysis and modeling resuls followed this process:

1. First, we read in our combined_hashtags.csv file with the combined hashtags of both **#trumptrain2020** and **#bidenharris2020** for a combined 2000 TikToks. Then, we will remove all of the stop words using Porter Stemmer for the most accurate keywords that may influence the model Trump and Biden in our descriptions.

2. Then, we will use a **Gaussian Classifier** and the **Naive Bayes algorithm** to train our model with the descriptions from each TikTok. This will be done by training our model on **80%** of the TikToks and use the remaining **20%** to test how accurate the model is.

```
df_combined_training_testing_data = pd.read_csv('combined_hashtags.csv')
nltk.download('stopwords')
training = []

# removing stop words and punctuation from descriptions of TikTok videos
for i in range(0, 2000):
    review = re.sub('[^a-zA-Z]', ' ', df_combined_training_testing_data['desc'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    training.append(review)

cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(training).toarray()
y = df_combined_training_testing_data['support']

# setting 80% of the data for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# creating our Gaussian model for training and testing
classifier = GaussianNB()
classifier.fit(X_train, y_train)
y_pred_NB_train = classifier.predict(X_train)
y_pred_NB_test = classifier.predict(X_test)

# output of training along with returned accuracy values.
confusionsTraining = confusion_matrix(y_train, y_pred_NB_train)
falseNegativeTraining = confusionsTraining[1][0] 
trueNegativeTraining = confusionsTraining[1][1]
falsePositiveTraining = confusionsTraining[0][1]
truePositiveTraining = confusionsTraining[0][0]

confusionsTesting = confusion_matrix(y_test, y_pred_NB_test)
falseNegativeTesting = confusionsTesting[1][0] 
trueNegativeTesting = confusionsTesting[1][1]
falsePositiveTesting = confusionsTesting[0][1]
truePositiveTesting = confusionsTesting[0][0]

confusions_normalized = confusionsTesting.astype('float') / confusionsTesting.sum(axis=1)[:, np.newaxis]
```

```
print(classification_report(y_train, y_pred_NB_train))
```

Upon re-running our model on the training data, we can see that our model reached an accuracy of about **90%.**

However the precision drops slightly due to our modeling mislabing some Trump-favored TikToks to be Biden-favored TikToks.

# **Ethics & Privacy**

The TikTok datasets we used in our project were retrieved using the Unofficial TikTok API developed by David Teather. The API could only retrieve existing videos and hashtags from publicly available TikTok user accounts. Thus, all of the data within the TikTok datasets are publicly available and would not raise any privacy concerns. However, we did not need any personally identifiable information from these accounts so we dropped them in our dataset. Therefore, our final TikTok datasets did not have values like TikTok creators' usernames, profile pictures, and personal bio. We also removed all ID's and links tied to the user accounts and videos. Removing these allows us to use our datasets without any privacy violations.

A potential privacy concern is if the TikTok data we scraped included a video that was later taken down or set to private. However, as mentioned previously, we removed every personally identifiable information and links to user accounts and videos. This way, even if videos are taken down later, there would be no way of tracking back to the original user or their post with our final data.

Another potential ethical concern is that our dataset could potentially misrepresent the primary demographic of TikTok users. We trust, however, in the prior research done by Jack Bandy mentioned above, which establishes that videos on TikTok become popular through its users. Since we scraped the most popular videos on TikTok through the Unofficial TikTok API, we can safely say that the data we obtained is representative of the primary TikTok demographic. By choosing to only scrape the most popular videos, we also avoid any bias in our dataset as the popular videos are more representative of the popular opinion.
